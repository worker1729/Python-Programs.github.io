```python
import pandas as pd
import matplotlib.pyplot as plt

#Create a DataFrame
names = ["Camile Winston", "Bernie Darnell", "Madeline Hollins", "Innocent Wyatt", "Freeman Gibbs"]
ages = [55, 23, 60, 63,37]

df = pd.DataFrame({"Name": names, "Age": ages})
print(df)

#Save our data in different files(CSV and Excel file)
save = df.to_csv("people.csv", index=False)
print("CSV file successfully created!")
save2 = df.to_excel('people.xlsx', sheet_name='Data', index=False, header=True)
print("Excel file successfully created!")
```
